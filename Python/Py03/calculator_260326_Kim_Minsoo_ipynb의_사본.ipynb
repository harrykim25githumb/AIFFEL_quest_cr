```python
cal_list = ['+', '-', '*', '/', '**']
import math
qst = "y"

while qst.lower() == "y":

    try:
        data_1, data_2 = [int(x) for x in input("연산할 정수 2개를 콤마로 구분하여 입력하시오 (ex) 3, 4:").split(",")]

    except ValueError as e1:
        print("잘못된 입력입니다. 정수를 입력해주세요: {}".format(e1))

    else:

        try:
            cal = str(input("연산자를 입력하시오 (+, -, *, /, ** 중 선택):"))

            if cal not in cal_list:
                raise ValueError("잘못된 입력입니다. 다시 입력해주세요")

            else:
                if cal == "+":
                    print("값:", data_1+data_2)
                if cal == "-":
                    print("값:", data_1-data_2)
                if cal == "*":
                    print("값:", data_1*data_2)
                if cal == "/":
                    print("값:", data_1/data_2)
                if cal == "**":
                    print("값:", math.pow(data_1, data_2))
                print("ok", cal)

        except ZeroDivisionError as e2:
            print("에러: {}".format(e2))

        except ValueError as e3:
            print("에러: {}".format(e3))


        ...

    finally:
        ...
        qst = str(input("계속 할 건가요 (Y/N)"))
```
