```python
''' Q1 ~ Q10 통합 '''

class Account():
    
    account_num = 0    #계좌 개수 초기값 설정
    account_list = []
    account_dict = {}
    deposit_log = []
    withdraw_log = []
    log_no = 1
    
    #Account 계좌 생성 모듈
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bank_name = 'SC은행'
        import random
        self.account_no = [random.randint(1,9) for i in range(11)]
        self.account_no2 = "".join(map(str, self.account_no))
        self.account_no3 = self.account_no2[0:3]+'-'+self.account_no2[3:5]+'-'+self.account_no2[5:]
        Account.account_num += 1
        self.dep_count = 0

        # self.account_info = {self.name, self.bank_name, self.account_no2, self.balance}
        Account.account_dict[self.name] = (self.bank_name, self.account_no2, self.balance)
        Account.account_list.append([self.name, self.bank_name, self.account_no2, self.balance])
        
    
    #계좌 객체 개수 출력
    def get_account_num():
        print("생성된 계좌 개수 : ", Account.account_num)

    #계좌 입금 모듈
    def deposit(self, amount):
        self.deposit_money = int(amount)
         
        if self.deposit_money >= 1:
            self.balance += self.deposit_money
            print(f"{self.name}님의 현재 계좌 잔액: {self.balance}원")
                       
            Account.account_dict[self.name] = (self.bank_name, self.account_no2, self.balance)
            Account.deposit_log.append((Account.log_no, '입금', self.name, self.deposit_money, self.balance)) 
            Account.log_no += 1

            self.dep_count +=1
            self.interest()
            
        else:
            print("입금은 1원 이상만 가능합니다")

    #계좌 출금 모듈
    def withdraw(self, amount):
        self.withdraw_money = int(amount)
        if self.withdraw_money > 0 and self.withdraw_money < self.balance :
            self.balance -= self.withdraw_money
            print(f"{self.name}님의 현재 계좌 잔액: {self.balance}원")
            Account.account_dict[self.name] = (self.bank_name, self.account_no2, self.balance)
            Account.withdraw_log.append((Account.log_no, '출금', self.name, self.withdraw_money, self.balance)) 
            Account.log_no += 1
        else:
            print("잔고이상으로 출금이 불가능합니다")

    #계좌 정보 출력
    def display_info(self):
        print(f'은행이름: {self.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_no3}, 잔고: {self.balance:,}원') 

    #이자 지급
    def interest(self):
        if self.dep_count == 5:
            self.int_money = self.balance * 0.01
            print(f"이자가 {self.int_money}원 지급되었습니다")
            self.balance += self.int_money
            self.dep_count = 0
            Account.account_dict[self.name] = (self.bank_name, self.account_no2, self.balance)
            Account.deposit_log.append((Account.log_no, '이자지급', self.name, self.int_money, self.balance))
            Account.log_no += 1
            return self.balance
            
    
    #잔고 100만원 이상 고객 정보
    vip_dict = {}
    def vip_list():
        for dict_name, dict_bal in Account.account_dict.items():            
            if dict_bal[2] >= 1000000:
               Account.vip_dict[dict_name] = dict_bal[2]
        print(Account.vip_dict)

    #입출금 기록
    def deposit_history():
        for i in range(len(Account.deposit_log)):
            print(Account.deposit_log[i])
    def withdraw_history():
        for i in range(len(Account.withdraw_log)):
            print(Account.withdraw_log[i])
        

a = Account('modu', 1)
b = Account('Lab', 0)
c = Account('Aiffel', 2000000)

Account.get_account_num()

a.deposit(10000)
a.withdraw(500)
a.deposit(100000)
a.deposit(100000)
a.deposit(100000)
a.deposit(1000000)
b.deposit(0)
b.withdraw(1000)
c.withdraw(500000)

a.display_info()
print(Account.account_dict)
Account.vip_list()
Account.deposit_history()
Account.withdraw_history()
```
